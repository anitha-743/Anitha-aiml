```python
"""line plot"""
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
plt.plot(x,y,marker='*',linestyle='-',color='purple',linewidth=2,markersize=4,label='sales')
plt.title("monthly sales trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""scatter plot"""
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
plt.scatter(x,y)
plt.show()
```

```python
"""size of points"""
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
sizes=[50,100,200,300,150]
plt.scatter(x,y,s=sizes)
plt.show()
```

```python
#plot multiple lines
import matplotlib.pyplot as plt
x=[1,2,3,4]
y1=[10,20,25,30]
y2=[40,50,60,70]
plt.plot(x,y1,label='Line 1',c='red')
plt.plot(x,y2,label='Line 2',c='purple')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(80)
y=np.random.rand(80)
sizes=10000*np.random.rand(80)
colors=np.random.rand(80)
plt.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.8)
plt.colorbar()
plt.title("rainbow scatter")
plt.xlabel("rain")
plt.ylabel("sun")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(10)
y=np.random.rand(10)
sizes=1000*np.random.rand(10)
colors=np.random.rand(10)
plt.scatter(x,y,s=sizes,c=colors,cmap='magma',alpha=0.8)
plt.colorbar()
plt.title("health maintaince")
plt.xlabel("fruits quantity")
plt.ylabel("vitamin quantity")
plt.show()
```
