```python
def average_yield(crop_yields):
  total_yield=sum( crop_yields)
  return total_yield/len(crop_yields)
crop_yields=[3000,6000,5000,7000,8000]
avg=average_yield(crop_yields)
print("average yield:",avg)
```

```python
def total_number_of_fields(crop_yields):
  return len(crop_yields)
crop_yields=[3000,6000,5000,7000,8000]
total_number_of_fields=total_number_of_fields(crop_yields)
print("total number of fields:",total_number_of_fields)
```

```python
def more_than_optimum_yield(crop_yields,threshold):
    more_than_optimum_yield=[]
    for value in crop_yields:
        if value>threshold:
            more_than_optimum_yield.append(value)
    return more_than_optimum_yield
crop_yields=[3000,6000,5000,7000,8000]
threshold=5000
print(more_than_optimum_yield(crop_yields,threshold))
```

```python
def yield_statistics(yields):
  print("total number of fields:",len(yields))
  print("maximum yield:",max(yields))
  print("minimum yield:",min(yields))
  print("average_yield:",avg)

crop_yields=[3000,6000,5000,7000,8000]
yield_statistics(crop_yields)
```
